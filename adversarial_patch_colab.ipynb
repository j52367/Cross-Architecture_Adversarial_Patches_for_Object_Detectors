## 1. COCO 데이터 준비

```python
DATA_DIR   = Path('data')
IMAGES_DIR = DATA_DIR / 'val2017'
ANNO_FILE  = DATA_DIR / 'annotations' / 'instances_val2017.json'
OUT_FILE   = DATA_DIR / 'person_annotations.json'

def download(url, dest):
    dest = Path(dest)
    dest.parent.mkdir(parents=True, exist_ok=True)
    if dest.exists():
        print(f'  이미 존재: {dest.name}')
        return
    print(f'  다운로드 중: {dest.name}')
    urllib.request.urlretrieve(url, dest)
    print('  완료')

def extract(zip_path, dest):
    print(f'  압축 해제: {Path(zip_path).name}')
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall(dest)

anno_zip = DATA_DIR / 'annotations.zip'
if not ANNO_FILE.exists():
    download('http://images.cocodataset.org/annotations/annotations_trainval2017.zip', anno_zip)
    extract(anno_zip, DATA_DIR)
    anno_zip.unlink(missing_ok=True)

images_zip = DATA_DIR / 'val2017.zip'
if not IMAGES_DIR.exists() or len(list(IMAGES_DIR.glob('*.jpg'))) < 100:
    download('http://images.cocodataset.org/zips/val2017.zip', images_zip)
    extract(images_zip, DATA_DIR)
    images_zip.unlink(missing_ok=True)

print('다운로드 완료')
```

```python
with open(ANNO_FILE) as f:
    coco = json.load(f)

person_cat_id = next(c['id'] for c in coco['categories'] if c['name'] == 'person')

image_id_to_bboxes = {}
for ann in coco['annotations']:
    if ann['category_id'] != person_cat_id:
        continue
    x, y, w, h = ann['bbox']
    if w < 30 or h < 60:
        continue
    image_id_to_bboxes.setdefault(ann['image_id'], []).append([x, y, x+w, y+h])

id_to_info = {img['id']: img for img in coco['images']}
records = []
for iid, bboxes in image_id_to_bboxes.items():
    info = id_to_info[iid]
    if not (IMAGES_DIR / info['file_name']).exists():
        continue
    records.append({'image_id': iid, 'file_name': info['file_name'],
                    'width': info['width'], 'height': info['height'], 'bboxes': bboxes})

with open(OUT_FILE, 'w') as f:
    json.dump(records, f)
print(f'person 이미지: {len(records)}장 저장 완료')
```

## 2. 공통 유틸 함수

```python
IMG_SIZE       = 640
PATCH_SIZE     = 300
PATCH_FRACTION = 0.35

def load_and_resize(file_name):
    img = Image.open(IMAGES_DIR / file_name).convert('RGB').resize((IMG_SIZE, IMG_SIZE))
    return torch.from_numpy(np.array(img)).permute(2, 0, 1).float() / 255.0

def scale_bboxes(bboxes, orig_w, orig_h):
    sx, sy = IMG_SIZE / orig_w, IMG_SIZE / orig_h
    return [[x1*sx, y1*sy, x2*sx, y2*sy] for x1, y1, x2, y2 in bboxes]

def apply_patch(img, patch, bboxes):
    """img: [3,H,W] GPU tensor  /  patch: [3,P,P] requires_grad"""
    H, W = img.shape[1], img.shape[2]
    mask   = torch.zeros(1, H, W, device=img.device)
    canvas = torch.zeros(3, H, W, device=img.device)
    for x1, y1, x2, y2 in bboxes:
        bw, bh = max(int(x2-x1), 1), max(int(y2-y1), 1)
        target = max(int(min(bw, bh) * PATCH_FRACTION), 16)
        p = F.interpolate(patch.unsqueeze(0), size=(target, target),
                          mode='bilinear', align_corners=False).squeeze(0)
        cx, cy = int((x1+x2)/2), int((y1+y2)/2)
        px1, py1 = max(0, cx-target//2), max(0, cy-target//2)
        px2, py2 = min(W, px1+target),   min(H, py1+target)
        ph, pw   = py2-py1, px2-px1
        if ph > 0 and pw > 0:
            mask[:, py1:py2, px1:px2]   = 1.0
            canvas[:, py1:py2, px1:px2] = p[:, :ph, :pw]
    return img * (1 - mask) + canvas * mask

def load_model(name):
    m  = YOLO(name)
    nn = m.model.to(DEVICE).eval()
    for p in nn.parameters():
        p.requires_grad_(False)
    return nn

def get_conf(nn_model, images):
    """YOLOv8 / RT-DETR 공용: 최대 confidence score 반환 (scalar)"""
    out   = nn_model(images)
    preds = out[0] if isinstance(out, (list, tuple)) else out
    if preds.dim() == 3:
        if preds.shape[1] < preds.shape[2]:   # [B, 4+nc, na]
            scores = preds[:, 4:, :].sigmoid().max(dim=1).values
        else:                                  # [B, na, 4+nc]
            scores = preds[:, :, 4:].sigmoid().max(dim=2).values
        return scores.max(dim=1).values.mean()
    raise ValueError(f'예상치 못한 출력 shape: {preds.shape}')

print('유틸 함수 로드 완료')
```

## 3. 모델 로드 및 출력 shape 확인

```python
print('모델 로드 중...')
yolo_nn   = load_model('yolov8n.pt')
rtdetr_nn = load_model('rtdetr-l.pt')

dummy = torch.zeros(1, 3, IMG_SIZE, IMG_SIZE, device=DEVICE)
with torch.no_grad():
    y_out = yolo_nn(dummy)
    r_out = rtdetr_nn(dummy)

for out, name in [(y_out, 'YOLOv8'), (r_out, 'RT-DETR')]:
    p = out[0] if isinstance(out, (list, tuple)) else out
    print(f'  {name}: {p.shape}')

print('확인 완료 — 위 shape이 dim=3이면 정상')
```

## 4. PGD 학습

`MODE` 를 바꿔서 세 가지 실험을 순서대로 실행하세요.
- `'yolo'` → Baseline A
- `'rtdetr'` → Baseline B
- `'ensemble'` → 제안 방법

```python
# ── 하이퍼파라미터 (T4 기준 1에폭 ≈ 1~2분) ─────────────
MODE       = 'ensemble'  # 'yolo' | 'rtdetr' | 'ensemble'
EPOCHS     = 30          # 본 실험: 30 / 빠른 테스트: 10
BATCH_SIZE = 8           # OOM 시 4로 줄이기
PGD_STEPS  = 10          # 본 실험: 10 / 빠른 테스트: 5
STEP_SIZE  = 2 / 255
ALPHA      = 0.5         # YOLO loss 가중치
BETA       = 0.5         # RT-DETR loss 가중치
MAX_IMAGES = None        # 본 실험: None (전체) / 빠른 테스트: 500
# ────────────────────────────────────────────────────────

Path('output').mkdir(exist_ok=True)

with open(OUT_FILE) as f:
    records = json.load(f)
random.shuffle(records)
if MAX_IMAGES:
    records = records[:MAX_IMAGES]
print(f'학습 이미지: {len(records)}장  |  모드: {MODE}')
```

```python
# 이미지를 RAM에 미리 올려둠 — 루프마다 디스크 읽기 방지
print('이미지 캐싱 중...')
cached_imgs   = []
cached_bboxes = []
for rec in tqdm(records, desc='캐싱'):
    cached_imgs.append(load_and_resize(rec['file_name']))  # CPU tensor
    cached_bboxes.append(scale_bboxes(rec['bboxes'], rec['width'], rec['height']))
print(f'캐싱 완료: {len(cached_imgs)}장')
```

```python
patch = torch.rand(3, PATCH_SIZE, PATCH_SIZE, device=DEVICE, requires_grad=True)
loss_history = []
indices = list(range(len(cached_imgs)))

for epoch in range(1, EPOCHS + 1):
    random.shuffle(indices)
    epoch_losses = []
    batches = [indices[i:i+BATCH_SIZE] for i in range(0, len(indices), BATCH_SIZE)]

    for idx_batch in tqdm(batches, desc=f'Epoch {epoch}/{EPOCHS}'):
        imgs     = [cached_imgs[i].to(DEVICE)   for i in idx_batch]
        bboxes_b = [cached_bboxes[i]             for i in idx_batch]

        for _ in range(PGD_STEPS):
            if patch.grad is not None:
                patch.grad.zero_()

            patched = torch.stack([apply_patch(img, patch, bx)
                                   for img, bx in zip(imgs, bboxes_b)])

            loss = torch.tensor(0.0, device=DEVICE)
            if MODE in ('yolo', 'ensemble'):
                loss = loss + ALPHA * get_conf(yolo_nn, patched)
            if MODE in ('rtdetr', 'ensemble'):
                loss = loss + BETA  * get_conf(rtdetr_nn, patched)

            loss.backward()

            with torch.no_grad():
                patch.data -= STEP_SIZE * patch.grad.sign()
                patch.data.clamp_(0, 1)

        epoch_losses.append(loss.item())

    avg = sum(epoch_losses) / len(epoch_losses)
    loss_history.append(avg)
    print(f'  Epoch {epoch}  avg loss: {avg:.4f}')
    torch.save(patch.detach().cpu(), f'output/{MODE}_patch_ep{epoch}.pt')

# 최종 저장
p_cpu = patch.detach().cpu()
torch.save(p_cpu, f'output/{MODE}_patch_final.pt')
arr = (p_cpu.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
Image.fromarray(arr).save(f'output/{MODE}_patch_final.png')
print(f'\n저장 완료: output/{MODE}_patch_final.pt')
```

```python
plt.figure(figsize=(7, 4))
plt.plot(range(1, len(loss_history)+1), loss_history, marker='o')
plt.title(f'{MODE} — Loss curve')
plt.xlabel('Epoch'); plt.ylabel('Avg confidence loss')
plt.grid(True); plt.tight_layout()
plt.savefig(f'output/{MODE}_loss.png')
plt.show()

patch_img = Image.open(f'output/{MODE}_patch_final.png')
plt.figure(figsize=(4, 4))
plt.imshow(patch_img); plt.title(f'{MODE} adversarial patch'); plt.axis('off')
plt.show()
```

## 5. α / β 가중치 탐색

앙상블에서 최적 α, β 조합을 찾습니다.  
각 조합을 3에폭만 빠르게 돌려서 loss로 비교합니다.

```python
AB_EPOCHS  = 3
AB_IMAGES  = 300
CONFIGS    = [(0.3, 0.7), (0.5, 0.5), (0.7, 0.3)]

ab_records  = records[:AB_IMAGES]
ab_imgs     = [cached_imgs[i]   for i in range(AB_IMAGES)]
ab_bboxes   = [cached_bboxes[i] for i in range(AB_IMAGES)]
ab_results  = {}

for alpha, beta in CONFIGS:
    label = f'alpha={alpha} beta={beta}'
    print(f'\n--- {label} ---')
    p = torch.rand(3, PATCH_SIZE, PATCH_SIZE, device=DEVICE, requires_grad=True)
    ep_losses = []

    for ep in range(AB_EPOCHS):
        idx2    = list(range(AB_IMAGES))
        random.shuffle(idx2)
        batches = [idx2[i:i+BATCH_SIZE] for i in range(0, AB_IMAGES, BATCH_SIZE)]
        step_losses = []

        for idx_b in tqdm(batches, desc=f'  ep{ep+1}', leave=False):
            imgs2 = [ab_imgs[i].to(DEVICE) for i in idx_b]
            bx2   = [ab_bboxes[i]          for i in idx_b]
            for _ in range(PGD_STEPS):
                if p.grad is not None: p.grad.zero_()
                pat2 = torch.stack([apply_patch(im, p, bx) for im, bx in zip(imgs2, bx2)])
                l    = alpha * get_conf(yolo_nn, pat2) + beta * get_conf(rtdetr_nn, pat2)
                l.backward()
                with torch.no_grad():
                    p.data -= STEP_SIZE * p.grad.sign()
                    p.data.clamp_(0, 1)
                step_losses.append(l.item())

        ep_avg = sum(step_losses) / len(step_losses)
        ep_losses.append(ep_avg)
        print(f'  ep{ep+1} loss: {ep_avg:.4f}')

    ab_results[label] = ep_losses[-1]
    torch.save(p.detach().cpu(), f'output/ensemble_{alpha}_{beta}_final.pt')

print('\n=== α/β 탐색 결과 (loss 낮을수록 공격력 강함) ===')
for k, v in sorted(ab_results.items(), key=lambda x: x[1]):
    print(f'  {k}: {v:.4f}')
```

## 6. 전이성 평가 (mAP Drop)

```python
def iou_score(b1, b2):
    xi1, yi1 = max(b1[0], b2[0]), max(b1[1], b2[1])
    xi2, yi2 = min(b1[2], b2[2]), min(b1[3], b2[3])
    inter = max(0, xi2-xi1) * max(0, yi2-yi1)
    union = (b1[2]-b1[0])*(b1[3]-b1[1]) + (b2[2]-b2[0])*(b2[3]-b2[1]) - inter
    return inter / union if union > 0 else 0.0

def compute_ap(rec, prec):
    return sum(
        (prec[rec >= t].max() if len(prec[rec >= t]) > 0 else 0)
        for t in np.arange(0, 1.1, 0.1)
    ) / 11

def evaluate_map(model_name, eval_records, patch_tensor=None):
    model = YOLO(model_name)
    all_tp, all_fp, total_gt = [], [], 0

    for rec in tqdm(eval_records, desc=f'  {model_name}', leave=False):
        img    = load_and_resize(rec['file_name']).to(DEVICE)
        bboxes = scale_bboxes(rec['bboxes'], rec['width'], rec['height'])
        total_gt += len(bboxes)

        if patch_tensor is not None:
            with torch.no_grad():
                img = apply_patch(img, patch_tensor.to(DEVICE), bboxes)

        img_np = (img.cpu().permute(1, 2, 0).clamp(0,1).numpy() * 255).astype(np.uint8)
        res    = model.predict(img_np, conf=0.25, iou=0.45, classes=[0], verbose=False)
        boxes  = res[0].boxes
        pred_boxes = boxes.xyxy.cpu().numpy() if len(boxes) > 0 else np.zeros((0, 4))
        pred_confs = boxes.conf.cpu().numpy() if len(boxes) > 0 else np.zeros(0)

        matched = [False] * len(bboxes)
        for pb, _ in sorted(zip(pred_boxes, pred_confs), key=lambda x: -x[1]):
            hit = next((i for i, gb in enumerate(bboxes)
                        if not matched[i] and iou_score(pb, gb) >= 0.5), None)
            all_tp.append(1 if hit is not None else 0)
            all_fp.append(0 if hit is not None else 1)
            if hit is not None:
                matched[hit] = True

    if not all_tp:
        return 0.0
    tp_c = np.cumsum(all_tp)
    fp_c = np.cumsum(all_fp)
    return compute_ap(tp_c / (total_gt + 1e-8), tp_c / (tp_c + fp_c + 1e-8))

print('평가 함수 로드 완료')
```

```python
EVAL_N       = 300
eval_records = records[:EVAL_N]

patches_to_eval = {
    'Clean':                       None,
    'Baseline A (YOLOv8 단독)':    'output/yolo_patch_final.pt',
    'Baseline B (RT-DETR 단독)':   'output/rtdetr_patch_final.pt',
    '제안 방법 (Ensemble)':        'output/ensemble_patch_final.pt',
}
eval_models  = ['yolov8n.pt', 'rtdetr-l.pt', 'yolov10n.pt']
results_table = {}

for patch_label, patch_path in patches_to_eval.items():
    if patch_path and not Path(patch_path).exists():
        print(f'스킵 (파일 없음): {patch_label}')
        continue
    patch_t = torch.load(patch_path, map_location=DEVICE) if patch_path else None
    print(f'\n[{patch_label}]')
    results_table[patch_label] = {}
    for m in eval_models:
        try:
            ap = evaluate_map(m, eval_records, patch_t)
            results_table[patch_label][m] = ap
            print(f'  {m:<15} AP={ap:.3f}')
        except Exception as e:
            print(f'  {m:<15} 에러: {e}')
```

```python
# 결과표 출력
clean = results_table.get('Clean', {})
col_w = 18

print('\n' + '='*72)
header = f'{"패치":<26}' + ''.join(f'{m.split(".")[0]:>{col_w}}' for m in eval_models)
print(header)
print('-'*72)

for label, row in results_table.items():
    line = f'{label:<26}'
    for m in eval_models:
        ap       = row.get(m, float('nan'))
        drop     = clean.get(m, ap) - ap
        cell     = f'{ap:.3f}({drop:+.3f})'
        line    += f'{cell:>{col_w}}'
    print(line)

print('='*72)
print('괄호: mAP drop  /  yolov10n = unseen 모델 (전이성 지표)')
```
